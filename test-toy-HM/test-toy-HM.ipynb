# terms
$\Lambda$ : Hyperparameter \
$\theta$ : Param of single distribution \
$\mu$ : Mean of the distribution \
$\sigma$ : variance of distribution \
$X$ : DATA \
$Z$: Evidence

For changing distribution of $\lambda$ https://docs.scipy.org/doc/scipy/reference/stats.html

Some helpful resources : https://www.pymc.io/projects/docs/en/stable/learn/core_notebooks/model_comparison.html, https://docs.exoplanet.codes/en/latest/tutorials/intro-to-pymc3/, https://boyangzhao.github.io/posts/mcmc-bayesian-inference


Solving $ P(\Lambda|X,\mu)$ using Bayes th. 

$ P(\Lambda|X,\mu) =\frac{P(X,\mu|\Lambda)\pi(\Lambda)}{Z}=\frac{P(X|\mu,\Lambda)\pi(\mu|\Lambda)\pi(\Lambda)}{Z}$

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import arviz as az
```

# Case 1 : $\Lambda$ is a Gaussian

```python
LEN_DATA = 500 # size of X
LEN_HYPERPARAM = 500 # mu in this case 
LAMBDA = st.norm(loc=3, scale=1).rvs(size=LEN_HYPERPARAM) # hyperparameter is governed by a gaussian function

plt.hist(LAMBDA, bins=50, fill=False, histtype='step')
plt.title(r'PDF$(\Lambda)=\mathcal{N}(3,1)$')
```

```python
X = np.empty((LEN_HYPERPARAM, LEN_DATA))
for i in range(LEN_HYPERPARAM):
    
    # Note here the std. dev of the X's is small. Large std. dev would lead
    # to incorrect estimation of "MU". For eg. std. dev of 1 would lead to 
    # MU ~ 1.5.
    
    X[i] = st.norm(loc=LAMBDA[i], scale=.05).rvs(size=LEN_DATA) 
```

```python
# large number of data sets can lead to a messy plot
for i in range(LEN_HYPERPARAM):
    plt.hist(X[i], fill=False, histtype='step', label=i)
#plt.legend()
plt.title(r'PDF$(X)=\mathcal{N}(\mu(\Lambda),1)$')
```

## USING MCMC (Metropolis Hasting algo)

```python
import pymc3 as pm
```

```python
with pm.Model() as model:

    hyperprior = pm.Normal('MU', mu=0, sigma=1)  # hyperprior
    prior = pm.Normal('mu', mu=hyperprior, sigma=.05)  # prior draws data points from hyperprior
    obs = pm.Normal('obs', mu=prior, sigma=1, observed=X)  # likelihood draws data points from prior
    step = pm.Metropolis()

    # sample with 3 independent Markov chains
    trace = pm.sample(draws=1000, chains=2, step=step, return_inferencedata=True)
```

```python
az.plot_trace(trace, var_names=('MU'))
```

```python
az.summary(trace, var_names=["MU"])["mean"]
```

# Case 2: $\Lambda$ is a lognorm

```python
LEN_DATA = 500 # size of X
LEN_HYPERPARAM = 500 
LAMBDA = st.lognorm(1).rvs(LEN_HYPERPARAM)
X = np.empty((LEN_HYPERPARAM, LEN_DATA))
```

```python
plt.hist(LAMBDA, bins=50, fill=False, histtype='step')
plt.title(r'$\Lambda$')
```

```python
X = np.empty((LEN_HYPERPARAM, LEN_DATA))
for i in range(LEN_HYPERPARAM):
    X[i] = st.norm(loc=LAMBDA[i], scale=0.5).rvs(size=LEN_DATA)
```

```python
# large number of data sets can lead to a messy plot
for i in range(LEN_HYPERPARAM):
    plt.hist(X[i], fill=False, histtype='step', label=i)
#plt.legend()
```

```python
import pymc3 as pm
```

```python
with pm.Model() as model:
    
    hyperprior = pm.LogNormal('MU', mu=0, sigma=1) # hyperprior
    prior = pm.Normal('mu', mu=hyperprior, sigma=0.5)  # prior draws data points from hyperprior
    obs = pm.Normal('obs', mu=prior, sigma=1, observed=X)  # likelihood draws data points from prior
    step = pm.Metropolis()
 
    # sample with 3 independent Markov chains
    trace = pm.sample(draws=1000, chains=3, step=step, return_inferencedata=True)
```

```python
az.plot_trace(trace, var_names=('MU'))
```

```python
az.summary(trace, var_names=["MU"])["mean"]
```

# Varying 2 parameters

```python
LEN_DATA = 500 # size of X
LEN_HYPERPARAM = 500 # mu in this case 
LAMBDA = st.gumbel_l(loc=2, scale=.5).rvs(size=LEN_HYPERPARAM) 
plt.hist(LAMBDA, bins=50, fill=False, histtype='step')
plt.title(r'PDF$(\Lambda)=\tilde{\mathcal{N}}(2,0.5)$')
```

```python
X = np.empty((LEN_HYPERPARAM, LEN_DATA))
for i in range(LEN_HYPERPARAM):
    X[i] = st.norm(loc=LAMBDA[i], scale=.5).rvs(size=LEN_DATA)
```

```python
# large number of data sets can lead to a messy plot
for i in range(LEN_HYPERPARAM):
    plt.hist(X[i], fill=False, histtype='step', label=i)
#plt.legend()
plt.title(r'PDF$(X)=\mathcal{N}(\mu(\Lambda),1)$')
```

```python
import pymc3 as pm
```

```python
with pm.Model() as model:

    MU = pm.Uniform('MU', lower=0, upper=5)
    SIGMA = pm.Uniform('SIGMA', lower=0, upper=1)
    hyperprior = pm.SkewNormal('hP', mu=MU, sigma=SIGMA)  # hyperprior
    prior = pm.Normal('mu', mu=hyperprior, sigma=1)  # prior draws data points from hyperprior
    obs = pm.Normal('obs', mu=prior, sigma=.5, observed=X)  # likelihood draws data points from prior
    step = pm.Metropolis()

    # sample with 3 independent Markov chains
    trace = pm.sample(draws=1500, chains=2, step=step, return_inferencedata=True)
```

```python
az.plot_trace(trace, var_names=('MU', 'SIGMA'))
# the reason I think this does not looks like a nice Gaussian PDF because this has been fitted with a different distribution
```

```python
az.summary(trace, var_names=["MU", "SIGMA"])["mean"]
```

```python
import corner
_ = corner.corner(trace, var_names=["MU", "SIGMA"], truths=dict(MU=2, SIGMA=.5))
```

# Varying multiple parameters

```python
LEN_DATA = 500 # size of X
LEN_HYPERPARAM = 500 # mu in this case 
LAMBDA = st.norm(loc=3, scale=.5).rvs(size=LEN_HYPERPARAM)*st.norm(loc=5, scale=1).rvs(size=LEN_HYPERPARAM)
plt.hist(LAMBDA, bins=50, fill=False, histtype='step')
plt.title(r'PDF$(\Lambda)=\mathcal{N}(3,0.5)*\mathcal{N}(5,1)$')
```

```python
X = np.empty((LEN_HYPERPARAM, LEN_DATA))
for i in range(LEN_HYPERPARAM):
    X[i] = (st.norm(loc=LAMBDA[i], scale=.5).rvs(size=LEN_DATA))
```

```python
# large number of data sets can lead to a messy plot
for i in range(LEN_HYPERPARAM):
    plt.hist(X[i], fill=False, histtype='step', label=i)
#plt.legend()
plt.title(r'PDF$(X)=\mathcal{N}(\mu(\Lambda),1)$')
```

```python
import pymc3 as pm
```

```python
with pm.Model() as model:

    MU = pm.Uniform('MU', lower=10, upper=30)
    SIGMA = pm.Uniform('SIGMA', lower=5, upper=10)
    hyperprior = pm.Normal('hP', mu=MU, sigma=SIGMA)  # hyperprior
    prior = pm.Normal('mu', mu=hyperprior, sigma=1)  # prior draws data points from hyperprior
    obs = pm.Normal('obs', mu=prior, sigma=.5, observed=X)  # likelihood draws data points from prior
    step = pm.Metropolis()

    # sample with 3 independent Markov chains
    trace = pm.sample(draws=500, chains=5, step=step, return_inferencedata=True)
```

```python
az.plot_trace(trace, var_names=('MU', 'SIGMA'))
```

```python
az.summary(trace, var_names=["MU", "SIGMA"])["mean"]
```

```python
import corner
_ = corner.corner(trace, var_names=["MU", "SIGMA"], truths=dict(MU=15, SIGMA=.5))
```
